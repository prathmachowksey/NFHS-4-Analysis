# 2. Clustering Analysis to identify factors that affect child stunting

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./../Data/working_data/data_attributes_removed_5_5.csv')
df.drop(df.columns[[0]], axis = 1, inplace = True) 
df
```

```python

df.columns
```

## 2.1 Selecting columns relevant to child stunting

```python
df = df[attributes_stunting]
```

```python
df.columns
```

## 2.2 Z-Score Normalisation - Required for clustering

```python
df_normalised = (df - df.mean())/df.std()
```

```python
df_normalised
```

## 2.3 clustering on the basis of sanitation facilities in household

```python

df_cluster_1 = df_normalised[['8. Households using improved sanitation facility2 (%)'] ]
```

```python
df_cluster_1
```

```python
from sklearn.cluster import KMeans
df_cluster_1_numpy = df_cluster_1.values
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_cluster_1_numpy)
cluster_labels = kmeans.labels_.tolist()
```

```python
df_cluster_1['cluster labels'] = cluster_labels
```

```python
df_cluster_1
```

```python
df_discrete = pd.read_csv('./../Data/working_data/data_attributes_removed_5_5.csv')
df_discrete.drop(df_discrete.columns[[0]], axis = 1, inplace = True) 
```

```python
stunting_discrete = df_discrete['68. Children under 5 years who are stunted11 (%)'].tolist()
```

```python
df_cluster_1['68. Children under 5 years who are stunted11 (%)'] = stunting_discrete
```

```python
df_cluster_1
```

```python
kmeans.cluster_centers_
```

**The cluster centers indicate that the clusters 0 , 1 and 2 correspond to districts having low , high and medium  % households having improved sanitation facilities**

```python
dfC0 = df_cluster_1[df_cluster_1['cluster labels']==0]
dfC1 = df_cluster_1[df_cluster_1['cluster labels']==1]
dfC2 = df_cluster_1[df_cluster_1['cluster labels']==2]

l,m,h=0,0,0

l = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC0.shape[0]
m = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC0.shape[0]
h = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC0.shape[0]
print("C0")
print(l)
print(m)
print(h)

l = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC1.shape[0]
m = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC1.shape[0]
h = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC1.shape[0]
print("C1")
print(l)
print(m)
print(h)

l = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC2.shape[0]
m = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC2.shape[0]
h = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC2.shape[0]
print("C2")
print(l)
print(m)
print(h)
```

It is clear of all the districts that belong to cluster 0 (corresponding to low % of households having improved santitation facilities ), 62.7% have 'high' stunting rates, and ONLY 7.5% have 'low' stunting rates. On the other hand, of all the points that belong to  cluster 1,(corresponding to high % of households having improved santitation facilities), 65.3% have 'low' stunting rates and ONLY 6.2% have 'high' % of children that are stunted .

**We can therefore conclude that children with access to improved sanitation facilities are at lower risk of stunting**

## 2.4 clustering on the basis of breastfeeding practises

```python

df_cluster_1 = df_normalised[['62. Children under age 3 years breastfed within one hour of birth8 (%)'] ]
```

```python
from sklearn.cluster import KMeans
df_cluster_1_numpy = df_cluster_1.values
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_cluster_1_numpy)
cluster_labels = kmeans.labels_.tolist()
```

```python
df_cluster_1
```

```python
df_discrete = pd.read_csv('./../Data/working_data/equal_frequency_discretised_data_7.csv')
df_discrete.drop(df_discrete.columns[[0]], axis = 1, inplace = True) 
df_discrete
```

```python
df_cluster_1
```

```python
kmeans.cluster_centers_
```

**The cluster centers indicate the cluster 0 , 1 and 2 correspond to data points having low , medium and high (ie, increasing) % of children who were breastfed within an hour of birth**

```python
dfC0 = df_cluster_1[df_cluster_1['cluster labels']==0]
dfC1 = df_cluster_1[df_cluster_1['cluster labels']==1]
dfC2 = df_cluster_1[df_cluster_1['cluster labels']==2]

l,m,h=0,0,0

l = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC0.shape[0]
m = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC0.shape[0]
h = dfC0[dfC0['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC0.shape[0]
print("C0")
print(l)
print(m)
print(h)

l = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC1.shape[0]
m = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC1.shape[0]
h = dfC1[dfC1['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC1.shape[0]
print("C1")
print(l)
print(m)
print(h)

l = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='low'].shape[0]/dfC2.shape[0]
m = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='medium'].shape[0]/dfC2.shape[0]
h = dfC2[dfC2['68. Children under 5 years who are stunted11 (%)']=='high'].shape[0]/dfC2.shape[0]
print("C2")
print(l)
print(m)
print(h)
```

**It is clear of all the points that belong to cluster 0 (corresponding to low % of children breastfed within an hour of birth), 47% have 'high' stunting rates. On the other hand, of all the points that belong to  cluster 2,(corresponding to low % of children breastfed within an hour of birth), 44% have 'low' stunting rates.**

**We can therefore conclude that early initiation to breastfeeding can prevent stunting in children.**
